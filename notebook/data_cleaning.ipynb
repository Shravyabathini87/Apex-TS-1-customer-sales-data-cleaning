```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw_data.csv")

df
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()

df
```

```python

df["Email"] = df["Email"].fillna("Not Available")

df = df.dropna(subset=["DateOfBirth"])

df
```

```python
df["TransactionDate"] = pd.to_datetime(
    df["TransactionDate"],
    errors="coerce",
    dayfirst=True
)

df["DateOfBirth"] = pd.to_datetime(df["DateOfBirth"])

df
```

```python
today = pd.Timestamp("today")

df["CustomerAge"] = (
    (today - df["DateOfBirth"]).dt.days // 365
)

df
```

```python
def product_category(p):
    if p in ["Laptop", "Mobile", "Tablet"]:
        return "Electronics"
    else:
        return "Accessories"

df["ProductCategory"] = df["Product"].apply(product_category)

df
```

```python
df
```

```python
df.to_csv("../data/cleaned_data.csv", index=False)

print("Cleaned dataset saved successfully!")
```
